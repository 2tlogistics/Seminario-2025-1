```python
def rotacion_inventario(cogs, inventario_promedio):
    return cogs / inventario_promedio

cogs = 500000
inventario_promedio = 100000
print(f"Rotación de inventario: {rotacion_inventario(cogs, inventario_promedio)}")
```

```python
def fill_rate(unidades_entregadas, unidades_solicitadas):
    return (unidades_entregadas / unidades_solicitadas) * 100

unidades_solicitadas = 1000
unidades_entregadas = 950
print(f"Fill Rate: {fill_rate(unidades_entregadas, unidades_solicitadas):.2f}%")
```

```python
def otif(pedidos_otif, total_pedidos):
    return (pedidos_otif / total_pedidos) * 100

pedidos_otif = 180
total_pedidos = 200
print(f"OTIF: {otif(pedidos_otif, total_pedidos):.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# Creación de dataset fijo
data = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'],
    'Rotación_Inventario': [4.2, 4.5, 4.8, 5.1, 5.3, 5.0],
    'OTIF': [0.85, 0.88, 0.90, 0.92, 0.91, 0.93],
    'Tiempo_Muerto_Horas': [40, 35, 30, 28, 25, 22],
    'Disponibilidad_Operacional': [0.88, 0.89, 0.90, 0.91, 0.92, 0.93],
    'Costo_Transporte_Unidad': [4.0, 4.1, 4.0, 3.9, 3.8, 3.7],
    'COGS': [400, 410, 420, 430, 440, 450],
    'Costo_Capital': 0.08, # 8% anual
    'Fill_Rate': [0.92, 0.93, 0.94, 0.95, 0.96, 0.97],
    'Beneficio_Operativo': [150, 155, 160, 165, 170, 175],
    'Costo_Mantenimiento': [40, 42, 38, 36, 35, 34],
    'Activos_Logísticos': [600, 610, 620, 630, 640, 650]
}

df = pd.DataFrame(data)
```

Cálculo de los KPIs

```python
# 1. Eficiencia Integrada Logística-Mantenimiento (EILM)
df['EILM'] = (df['Rotación_Inventario'] * df['OTIF']) / (df['Tiempo_Muerto_Horas'] / 100) * df['Disponibilidad_Operacional']

# 2. Costo Logístico-Financiero por Unidad (CLFU)
df['CLFU'] = (df['Costo_Transporte_Unidad'] + (df['COGS'] * df['Costo_Capital'] / 12)) / df['Fill_Rate']

# 3. ROI Logístico Ajustado (ROILA)
df['Stockout_Rate'] = 1 - df['Fill_Rate']
df['ROILA'] = ((df['Beneficio_Operativo'] - (df['Costo_Mantenimiento'] * df['Stockout_Rate'])) / df['Activos_Logísticos']) * 100

# Mostrar resultados
print(df[['Mes', 'EILM', 'CLFU', 'ROILA']].round(2))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# Verificar estilos disponibles
print("Estilos disponibles:", plt.style.available)

# Usaremos 'ggplot' que es comúnmente disponible
plt.style.use('ggplot')

# Creación de dataset fijo
data = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'],
    'Rotación_Inventario': [4.2, 4.5, 4.8, 5.1, 5.3, 5.0],
    'OTIF': [0.85, 0.88, 0.90, 0.92, 0.91, 0.93],
    'Tiempo_Muerto_Horas': [40, 35, 30, 28, 25, 22],
    'Disponibilidad_Operacional': [0.88, 0.89, 0.90, 0.91, 0.92, 0.93],
    'Costo_Transporte_Unidad': [4.0, 4.1, 4.0, 3.9, 3.8, 3.7],
    'COGS': [400, 410, 420, 430, 440, 450],
    'Costo_Capital': 0.08, # 8% anual
    'Fill_Rate': [0.92, 0.93, 0.94, 0.95, 0.96, 0.97],
    'Beneficio_Operativo': [150, 155, 160, 165, 170, 175],
    'Costo_Mantenimiento': [40, 42, 38, 36, 35, 34],
    'Activos_Logísticos': [600, 610, 620, 630, 640, 650]
}

df = pd.DataFrame(data)

# Cálculo de KPIs híbridos
df['EILM'] = (df['Rotación_Inventario'] * df['OTIF']) / (df['Tiempo_Muerto_Horas'] / 100) * df['Disponibilidad_Operacional']
df['CLFU'] = (df['Costo_Transporte_Unidad'] + (df['COGS'] * df['Costo_Capital'] / 12)) / df['Fill_Rate']
df['Stockout_Rate'] = 1 - df['Fill_Rate']
df['ROILA'] = ((df['Beneficio_Operativo'] - (df['Costo_Mantenimiento'] * df['Stockout_Rate'])) / df['Activos_Logísticos']) * 100

# Función para crear el dashboard corregida
def create_simple_dashboard(df):
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

    # Gráfico EILM
    ax1.plot(df['Mes'], df['EILM'], marker='o', color='blue')
    ax1.axhline(y=15, color='green', linestyle='--', label='Meta Óptima')
    ax1.set_title('Eficiencia Integrada (EILM)')
    ax1.set_ylabel('Puntaje EILM')
    ax1.legend()

    # Gráfico CLFU
    ax2.bar(df['Mes'], df['CLFU'], color='orange')
    ax2.axhline(y=6.5, color='red', linestyle='--', label='Límite Aceptable')
    ax2.set_title('Costo Logístico-Financiero por Unidad (CLFU)')
    ax2.set_ylabel('Dólares')
    ax2.legend()

    # Gráfico ROILA
    ax3.plot(df['Mes'], df['ROILA'], marker='s', color='green')
    ax3.axhline(y=25, color='blue', linestyle='--', label='Meta Óptima')
    ax3.set_title('ROI Logístico Ajustado (ROILA)')
    ax3.set_ylabel('Porcentaje')
    ax3.legend()

    plt.tight_layout()
    return fig

# Generar y mostrar dashboard
dashboard = create_simple_dashboard(df)
plt.show()

# Exportar a PDF
def export_analysis(df):
    with PdfPages('Analisis_KPIs_Hibridos.pdf') as pdf:
        # Portada
        plt.figure(figsize=(8.5, 11))
        plt.axis('off')
        plt.text(0.5, 0.7, 'Análisis de KPIs Híbridos',
                ha='center', va='center', fontsize=20)
        plt.text(0.5, 0.5, 'Ejemplo Práctico con Datos Fijos\n\nTechLogistic S.A.',
                ha='center', va='center', fontsize=14)
        pdf.savefig(bbox_inches='tight')
        plt.close()

        # Dashboard
        dashboard = create_simple_dashboard(df)
        pdf.savefig(dashboard, bbox_inches='tight')
        plt.close()

        # Análisis
        plt.figure(figsize=(8.5, 11))
        plt.axis('off')

        analysis_text = """
        ANÁLISIS DETALLADO

        1. EILM (Eficiencia Integrada Logística-Mantenimiento):
        - Rango óptimo: >15
        - Valor actual: {:.2f} (Junio)
        - Tendencia: Mejora constante

        2. CLFU (Costo Logístico-Financiero por Unidad):
        - Rango óptimo: <$6.50
        - Valor actual: ${:.2f} (Junio)
        - Tendencia: Reducción constante

        3. ROILA (ROI Logístico Ajustado):
        - Rango óptimo: >25%
        - Valor actual: {:.2f}% (Junio)
        - Tendencia: Crecimiento sostenido
        """.format(df['EILM'].max(), df['CLFU'].min(), df['ROILA'].max())

        plt.text(0.1, 0.5, analysis_text, fontsize=12)
        pdf.savefig(bbox_inches='tight')
        plt.close()

export_analysis(df)
print("Análisis exportado como 'Analisis_KPIs_Hibridos.pdf'")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# 1. Creación del dataset fijo
data = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'],
    'Rotación_Inventario': [4.2, 4.5, 4.8, 5.1, 5.3, 5.0],
    'OTIF': [0.85, 0.88, 0.90, 0.92, 0.91, 0.93],
    'Tiempo_Muerto_Horas': [40, 35, 30, 28, 25, 22],
    'Disponibilidad_Operacional': [0.88, 0.89, 0.90, 0.91, 0.92, 0.93],
    'Costo_Transporte_Unidad': [4.0, 4.1, 4.0, 3.9, 3.8, 3.7],
    'COGS': [400, 410, 420, 430, 440, 450],
    'Costo_Capital': 0.08,
    'Fill_Rate': [0.92, 0.93, 0.94, 0.95, 0.96, 0.97],
    'Beneficio_Operativo': [150, 155, 160, 165, 170, 175],
    'Costo_Mantenimiento': [40, 42, 38, 36, 35, 34],
    'Activos_Logísticos': [600, 610, 620, 630, 640, 650]
}

df = pd.DataFrame(data)

# 2. Cálculo de KPIs híbridos
df['EILM'] = (df['Rotación_Inventario'] * df['OTIF']) / (df['Tiempo_Muerto_Horas'] / 100) * df['Disponibilidad_Operacional']
df['CLFU'] = (df['Costo_Transporte_Unidad'] + (df['COGS'] * df['Costo_Capital'] / 12)) / df['Fill_Rate']
df['Stockout_Rate'] = 1 - df['Fill_Rate']
df['ROILA'] = ((df['Beneficio_Operativo'] - (df['Costo_Mantenimiento'] * df['Stockout_Rate'])) / df['Activos_Logísticos']) * 100

# 3. Función para crear el dashboard de gestión
def create_management_dashboard(dataframe):
    fig, ax = plt.subplots(figsize=(12, 6))

    # Configurar datos
    meses = dataframe['Mes']
    indicadores = {
        'EILM': (dataframe['EILM'], 15, 25, 'blue'),
        'CLFU': (dataframe['CLFU'], 6.5, 5.0, 'orange'),
        'ROILA': (dataframe['ROILA'], 25, 30, 'green')
    }

    # Crear gráfico
    width = 0.25
    x = np.arange(len(meses))

    for i, (kpi, (values, meta, objetivo, color)) in enumerate(indicadores.items()):
        ax.bar(x + i*width, values, width, label=kpi, color=color)
        ax.axhline(y=meta, color=color, linestyle='--', alpha=0.5)
        ax.axhline(y=objetivo, color=color, linestyle='-', alpha=0.3)

    # Personalizar gráfico
    ax.set_xticks(x + width)
    ax.set_xticklabels(meses)
    ax.set_title('KPIs vs Metas (Línea punteada = Meta, Línea sólida = Objetivo)')
    ax.legend()
    ax.grid(True, axis='y', alpha=0.3)

    plt.tight_layout()
    return fig

# 4. Generar y mostrar dashboard
dashboard = create_management_dashboard(df)  # Pasar el DataFrame definido
plt.show()

# 5. Exportar a PDF (opcional)
def export_to_pdf(fig):
    with PdfPages('dashboard_kpis.pdf') as pdf:
        pdf.savefig(fig, bbox_inches='tight')
    print("Dashboard exportado como 'dashboard_kpis.pdf'")

export_to_pdf(dashboard)
```
